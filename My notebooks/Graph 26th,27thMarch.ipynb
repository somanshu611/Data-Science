```python
# Scatter plot
# For finding correlation
df.plot.scatter(x='SepalLengthCm',y='PetalLengthCm')
# strong relationship
```

```python
df.plot.scatter(x='SepalLengthCm',y='PetalWidthCm')
# too scattered 
```

```python
df.plot.scatter(x='SepalLengthCm',y='PetalLengthCm',c='SepalWidthCm',figsize=(20,10),s=200)
# s = size of dots

```

```python
# you can also write 's' as a function of other variable.
# lIke this : 
df.plot.scatter(x='SepalLengthCm',y='PetalLengthCm',c='SepalWidthCm',figsize=(20,10),s=df['PetalWidthCm']*300)
```

```python
df.plot.hexbin(x='SepalLengthCm',y='PetalLengthCm',gridsize=10,C='PetalWidthCm')
```
