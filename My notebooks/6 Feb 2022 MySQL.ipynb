```python
import csv
with open('car.data') as f: 
    car_data = csv.reader(f, delimiter='\n')
    for i in car_data:
#         print(i)
        s = ""
        st = "'"
        for j in i[0]: 
            if j == ',' : 
                s+=st+"', '"
                st = ''
            else : 
                st+=j
        s+=st+"'"
        print(s)
        c.execute("insert into cardataset.car values ({})".format(s))
    mydb.commit()
```
